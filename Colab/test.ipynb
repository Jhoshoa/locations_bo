```python
# Import necessary libraries
import psycopg2
import pandas as pd

# Database connection parameters
db_params = {
    'dbname': 'locationdb',
    'user': 'usertest',
    'password': 'password',
    'host': 'localhost',
    'port': '5432'
}

# Function to connect to the database and fetch data
def fetch_data(query):
    try:
        # Establish the connection
        conn = psycopg2.connect(**db_params)
        # Create a cursor object
        cursor = conn.cursor()
        # Execute the query
        cursor.execute(query)
        # Fetch all results
        results = cursor.fetchall()
        # Get column names
        colnames = [desc[0] for desc in cursor.description]
        # Close the cursor and connection
        cursor.close()
        conn.close()
        # Return results as a DataFrame
        return pd.DataFrame(results, columns=colnames)
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

query = "SELECT * FROM tmp_recintosgeo;"
data = fetch_data(query)

columns_name = data.columns

## Departamentos
get_dep = data[['dep','nomdep']].drop_duplicates(subset='dep')

## Provincias
get_provincias = data[['dep', 'prov', 'nomprov']].drop_duplicates(subset='nomprov')

if data is not None:
    print(columns_name)
    # print(get_dep)
    print("Numero de Provincia: " + str(len(get_provincias)))
    print(get_provincias)
    print(data)
```
